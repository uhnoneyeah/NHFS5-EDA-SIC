```python
wide_df = df.pivot_table(
    index="state",
    columns="indicator",
    values="nfhs5_total",
    aggfunc="first"
).reset_index()

wide_df.head()
```

```python
print(wide_df.columns.tolist())
```

```python
analysis_df = wide_df[[
    "state",
    "88. Women who are overweight or obese (BMI ≥25.0 kg/m2)21 (%)",
    "89. Men who are overweight or obese (BMI ≥25.0 kg/m2) (%)",
    "92. Children age 6-59 months who are anaemic (<11.0 g/dl)22 (%)",
    "95. All women age 15-49 years who are anaemic22 (%)",
    "101. Blood sugar level - high or very high (>140 mg/dl) or taking medicine to control blood sugar level23 (%)",
    "104. Blood sugar level - high or very high (>140 mg/dl) or taking medicine to control blood sugar level23 (%)",
    "107. Elevated blood pressure (Systolic ≥140 mm of Hg and/or Diastolic ≥90 mm of Hg) or taking medicine to control blood pressure (%)",
    "110. Elevated blood pressure (Systolic ≥140 mm of Hg and/or Diastolic ≥90 mm of Hg) or taking medicine to control blood pressure (%)"
]]
```

```python
analysis_df.columns = [
    "State",
    "Obesity_Women",
    "Obesity_Men",
    "Anaemia_Children",
    "Anaemia_Women",
    "High_Blood_Sugar_Women",
    "High_Blood_Sugar_Men",
    "Hypertension_Women",
    "Hypertension_Men"
]
```

```python
numeric_cols = analysis_df.columns[1:]

for col in numeric_cols:
    analysis_df[col] = pd.to_numeric(analysis_df[col], errors="coerce")
```

```python
analysis_df["Gender_Obesity_Gap"] = (
    analysis_df["Obesity_Women"] -
    analysis_df["Obesity_Men"]
)
```

```python
analysis_df.head()
analysis_df.info()
```

```python
analysis_df
```

```python
correlation = analysis_df[
    ["Anaemia_Children", "Obesity_Men", "Obesity_Women"]
].corr(method="pearson")

print(correlation)
```

```python
plt.figure(figsize=(7,6))

sns.heatmap(
    correlation,
    annot=True,
    cmap="coolwarm",
    linewidths=0.5,
    fmt=".2f"
)

plt.title("Correlation between Childhood Anaemia and Adult Obesity")
plt.tight_layout()
plt.show()
```

```python
top10 = analysis_df.sort_values(
    by="Hypertension_Women",
    ascending=False
).head(10)

top10
```

```python
plt.figure(figsize=(12,6))

x = range(len(top10))
width = 0.4

plt.bar(
    [i-width/2 for i in x],
    top10["Hypertension_Men"],
    width=width,
    color="steelblue",
    label="Men"
)

plt.bar(
    [i+width/2 for i in x],
    top10["Hypertension_Women"],
    width=width,
    color="tomato",
    label="Women"
)

plt.xticks(
    x,
    top10["State"],
    rotation=45,
    ha="right"
)

plt.xlabel("State")
plt.ylabel("Hypertension (%)")
plt.title("Top 10 States with Highest Hypertension Rates")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```
